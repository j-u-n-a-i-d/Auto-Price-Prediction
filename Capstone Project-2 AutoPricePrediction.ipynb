# Domain Analysis
- symboling: a rating system used to assess the riskiness of a car, often used by insurance companies
- normalized_losses: a numerical value representing the average loss payment per insured vehicle year, normalized for all vehicles in a particular group
- make: the brand or manufacturer of the car
- fuel_type: the type of fuel the car uses (e.g. gasoline, diesel, electric)
- aspiration: whether the car has a turbocharger or supercharger
- num_of_doors: the number of doors on the car
- body_style: the body configuration of the car (e.g. sedan, coupe, hatchback)
- drive_wheels: the wheels that provide power to the car (e.g. front-wheel drive, rear-wheel drive, all-wheel drive)
- engine_location: the placement of the engine in the car (e.g. front, rear, mid)
- wheel_base: the distance between the centers of the front and rear wheels of the car
- length: the overall length of the car
- width: the overall width of the car
- height: the overall height of the car
- curb_weight: the weight of the car without any passengers or cargo
- engine_type: the type of engine the car has (e.g. inline, V-shaped, rotary)
- num_of_cylinders: the number of cylinders in the car's engine
- engine_size: the displacement of the engine in cubic centimeters (cc)
- fuel_system: the method used to deliver fuel to the engine (e.g. carburetor, fuel injection)
- bore: the diameter of each cylinder in the engine
- stroke: the distance the piston travels up and down in each cylinder
- compression_ratio: the ratio of the volume of the cylinder when the piston is at the bottom of its stroke (i.e. the volume of the combustion chamber) to the volume when the piston is at the top of its stroke
- horsepower: the power output of the engine in horsepower (hp)
- peak_rpm: the engine speed at which the maximum horsepower is reached
- city_mpg: the estimated fuel efficiency of the car in city driving conditions, measured in miles per gallon
- highway_mpg: the estimated fuel efficiency of the car in highway driving conditions, measured in miles per gallon
- price: the cost of the car, typically measured in a particular currency (e.g. US dollars)
